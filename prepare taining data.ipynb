```python
import pandas as pd 
```

```python
def selection (data,d_type):
    return pd.DataFrame(data.select_dtypes(include = d_type))
```

```python
df_item_cat = pd.read_csv('item_categories.csv')
```

```python
df_item_cat
```

```python
df_item = pd.read_csv('items.csv')
df_shops = pd.read_csv('shops.csv')
df_sales_train = pd.read_csv('sales_train.csv')
```

```python
 result1 = pd.merge(df_item_cat,df_item , how="inner", on=["item_category_id"])
```

```python
result2 = pd.merge(df_sales_train, result1, how="inner", on=["item_id"])
```

```python
result3 = pd.merge(df_shops, result2, how="inner", on=["shop_id"])
```

```python
data_for_preprocess=result3.drop(['item_id', 'shop_id','item_category_id','date'], axis=1)
```

```python
d_f=selection(data_for_preprocess,'float64')
d_int=selection(data_for_preprocess,'int64')
d_object=selection(data_for_preprocess,'object')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df1=pd.DataFrame()

for (columnName, columnData) in d_object.iteritems(): 
      df1[columnName] = le.fit_transform(d_object[columnName])
      df1.append(df1)
```

```python
df1
```

```python
frames = [df1,d_int,d_f]
X_TRAIN = pd.concat(frames,axis=1)
```

```python
X_TRAIN.head(3)
```

```python
X_TRAIN_f =pd.DataFrame(X_TRAIN.drop_duplicates())
```

```python
X_TRAIN_f
```

```python
X_TRAIN_f1=X_TRAIN_f.drop(["item_cnt_day"], axis = 1) 
```

```python
Y_TRAIN =pd.DataFrame(X_TRAIN_f["item_cnt_day"])
```

```python
X_TRAIN_f1.to_csv('train_x.csv')
Y_TRAIN.to_csv('train_y.csv')
```
